```python
df = pd.DataFrame(data3)

def convert_exponents_to_numeric(value):
    try:
        return float(value)
    except ValueError:
        return None

result_df['Rb (ohm cm2) '] = df['new2'].map(convert_exponents_to_numeric)
```

```python
interpolated_df = pd.DataFrame()

num_interpolated_rows = 10
for i in range(result_df.shape[0] - 1):
    start_row = result_df.iloc[i]
    end_row = result_df.iloc[i + 1]

    for j in range(num_interpolated_rows + 1):
        interpolation_factor = j / num_interpolated_rows
        interpolated_row = start_row + (end_row - start_row) * interpolation_factor
        interpolated_df = pd.concat([interpolated_df, pd.DataFrame(interpolated_row).transpose()], ignore_index=True)

result_df_interpolated = pd.concat([result_df, interpolated_df], ignore_index=True)

print(result_df_interpolated.head(20))
```

```python
result_df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error



features_columns = ['n.1', 'Rpor (ohm cm2) ', 'n.2', 'CPEpor', 'CPEb']
target_column = 'Rb (ohm cm2) '

selected_columns = features_columns + [target_column]
dataset_selected = result_df_interpolated [selected_columns]

train_data, test_data = train_test_split(dataset_selected, test_size=0.2, random_state=42)

X_train = train_data[features_columns]
y_train = train_data[target_column]

X_test = test_data[features_columns]
y_test = test_data[target_column]

model = LinearRegression()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

# Model fitting
lr = LinearRegression()
lr.fit(X_train, y_train)

# Coefficients of the 2nd degree polynomial equation
coef = lr.coef_
intercept = lr.intercept_
print("Coefficients:", coef)
print("Intercept:", intercept)

```

```python
feature_importance = permutation_importance(model, X_test, y_test, n_repeats=30, random_state=42)
sorted_idx = feature_importance.importances_mean.argsort()[::-1]

print("Feature Importance:")
for idx in sorted_idx:
    print(f"{features_columns[idx]}: {feature_importance.importances_mean[idx]}")
```

```python
print("Training set shapes:")
print("X_train:", X_train.shape)
print("y_train:", y_train.shape)

print("\nTesting set shapes:")
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```

```python
r_squared = r2_score(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f"R-squared on Test Data: {r_squared}")
print(f"Mean Squared Error on Test Data: {mse}")
print(f"Root Mean Squared Error on Test Data: {rmse}")
```

```python
residuals = y_test - predictions

plt.scatter(predictions, residuals)
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()

```
